### 5. Data Selection - Based on Conditional Filtering

```python
# Use this code block to select movies that have been released between 2010-2016, with a rating of less than 6.0, but generated significant revenue.
# For "significant revenue", filter for movies in the top 25% of revenue (use data['Revenue (Millions)'].quantile(0.75))

revenue = data['Revenue (Millions)'].quantile(0.75)
print(revenue)
filtered_movies = data[
    (data['Year'] >= 2010) &
    (data['Year'] <= 2016) &
    (data['Rating'] < 6.0) &
    (data['Revenue (Millions)'] > revenue)
]

print(filtered_movies)
```

```python
# Bonus Challenge: Try adjusting the quantile value to be more specific.
# Can you find movies with exactly the characteristics you're looking for by experimenting with different thresholds?

# Try different quantile thresholds (0.50, 0.60, 0.80, 0.90, etc.)
threshold = 0.85   

revenue_cutoff = data['Revenue (Millions)'].quantile(threshold)

filtered_movies = data[
    (data['Genre'] == 'Comedy') &
    (data['Revenue (Millions)'] > revenue_cutoff)
]

print("Quantile used:", threshold)
print("Revenue cutoff:", revenue_cutoff)
print(filtered_movies)
```
